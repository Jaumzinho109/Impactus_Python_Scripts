```python
#Aqui vou colocar retail sales excl motor vehicle junto com gráfico de retail sales normal



def plot_mom_2_bars(df, df1, titulo="Título aqui em", label1="Série 1", label2="Série 2"):
    df = df.tail(18)
    df1 = df1.tail(18)

    fig, ax = plt.subplots(figsize=(15,9))
    
    x = df.index
    width = 9
    
    bars1 = ax.bar(x - pd.Timedelta(days=width/2), df["Pct Change"], width=width, color="#082631", label=label1)
    bars2 = ax.bar(x + pd.Timedelta(days=width/2), df1["Pct Change"], width=width, color="#37A6D9", label=label2)

    ax.set_title("MoM % SA", fontsize=18, style='italic', pad=10)
    fig.suptitle(titulo, fontsize=28, fontweight='bold')

    ax.xaxis.set_major_locator(mdates.MonthLocator(bymonth=[2,4,6,8,10,12])) 
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_color('#d9d9d9')
    ax.spines["bottom"].set_color('#d9d9d9')

    ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0))
    ax.set_xlabel("Fonte: FRED | Impactus UFRJ", fontsize=18, labelpad=15)
    ax.axhline(0, color='black', lw=0.8)

    ax.legend(fontsize=18)

    for bar in bars1:
        height = bar.get_height()
        if height != 0:
            ax.annotate(f'{height:.1%}',
                        xy=(bar.get_x() + bar.get_width()/2, height),
                        xytext=(0, 3 if height > 0 else -10),
                        textcoords="offset points",
                        ha='center', va='bottom' if height > 0 else 'top',
                        fontsize=10)

    for bar in bars2:
        height = bar.get_height()
        if height != 0:
            ax.annotate(f'{height:.1%}',
                        xy=(bar.get_x() + bar.get_width()/2, height),
                        xytext=(0, 3 if height > 0 else -10),
                        textcoords="offset points",
                        ha='center', va='bottom' if height > 0 else 'top',
                        fontsize=10)

    fig.autofmt_xdate(rotation=0)
    plt.tight_layout()

    return fig






graf_retail_sales_2_bars = plot_mom_2_bars(retail_sales,retail_sales_excl_motor_vehicle, "Retail Sales", "Retail Sales", "Retail Sales Excl Motor Vehicle")
plt.show(graf_retail_sales_2_bars)
```

```python

def qoq(df, titulo="Título aqui em"):
    df = df.dropna()
    df = df.tail(12)
    df["Anualized"] = df["Pct Change"].apply(lambda x: (1 + x) ** 4 - 1)
    
    fig, ax = plt.subplots(figsize=(15, 9))
    bars = ax.bar(df.index, df["Anualized"], width=14, color="#082631")
    
    fig.suptitle(titulo, fontsize=28, fontweight='bold')
    ax.set_title("QoQ % SAAR SA", fontsize=18, style='italic', pad=10)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_color('#d9d9d9')
    ax.spines["bottom"].set_color('#d9d9d9')
    
    ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0))
    ax.set_xlabel("Fonte: FRED | Impactus UFRJ", fontsize=18, labelpad=15)
    ax.axhline(0, color='black', lw=0.8)

    for bar in bars:
        height = bar.get_height()
        if height >= 0:
            ax.text(bar.get_x() + bar.get_width() / 2, height + 0.0003,
                    f'{height:.1%}', ha='center', va='bottom', fontsize=14, color="#082631")
        else:
            ax.text(bar.get_x() + bar.get_width() / 2, height - 0.0003,
                    f'{height:.1%}', ha='center', va='top', fontsize=14, color="#082631")
    
    plt.tight_layout()
    return fig

```

```python
gci = fred.get_series("GCEC1")
fci = fred.get_series("FGCEC1")
rsci = fred.get_series("SLCEC1")
government_consumption_expenditures_and_inv = pd.DataFrame()
government_consumption_expenditures_and_inv["Government Consumption Expenditures and Gross Investment"] = pd.DataFrame(gci)
government_consumption_expenditures_and_inv["Real Federal Consumption Expenditures and Gross Investment"] = pd.DataFrame(fci)
government_consumption_expenditures_and_inv["Real State and Local Consumption Expenditures & Gross Investment"] = pd.DataFrame(rsci)
government_consumption_expenditures_and_inv["Government Pct Change"]= government_consumption_expenditures_and_inv["Government Consumption Expenditures and Gross Investment"].pct_change()
government_consumption_expenditures_and_inv["Real Federal Consumption Expenditures and Gross Investment Pct Change"]= government_consumption_expenditures_and_inv["Real Federal Consumption Expenditures and Gross Investment"].pct_change()
government_consumption_expenditures_and_inv["Real State and Local Consumption Expenditures & Gross Investment Pct Change"]= government_consumption_expenditures_and_inv["Real State and Local Consumption Expenditures & Gross Investment"].pct_change()

def qoq_grouped(df, titulo="Título aqui em"):
    df = df.dropna()
    df = df.tail(6)

    df["Anualized Gov"] = (1 + df["Government Pct Change"]) ** 4 - 1
    df["Anualized Fed"] = (1 + df["Real Federal Consumption Expenditures and Gross Investment Pct Change"]) ** 4 - 1
    df["Anualized State"] = (1 + df["Real State and Local Consumption Expenditures & Gross Investment Pct Change"]) ** 4 - 1

    fig, ax = plt.subplots(figsize=(15, 9))
    bar_width = 20  
    index = pd.to_datetime(df.index)

    ax.set_prop_cycle(color=["#082631", "#37A6D9", "#AFABAB"])

    bars1 = ax.bar(index - pd.Timedelta(days=bar_width), df["Anualized Gov"], width=bar_width, label='Total')
    bars2 = ax.bar(index, df["Anualized Fed"], width=bar_width, label='Federal')
    bars3 = ax.bar(index + pd.Timedelta(days=bar_width), df["Anualized State"], width=bar_width, label='State & Local')

    fig.suptitle(titulo, fontsize=28, fontweight='bold')
    ax.set_title("QoQ % SAAR SA", fontsize=18, style='italic', pad=10)

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_color('#d9d9d9')
    ax.spines["bottom"].set_color('#d9d9d9')

    ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1.0))
    ax.set_xlabel("Fonte: FRED | Impactus UFRJ", fontsize=18, labelpad=15)
    ax.axhline(0, color='black', lw=0.8)

    for bars in [bars1, bars2, bars3]:
        for bar in bars:
            height = bar.get_height()
            if pd.notna(height):
                if height >= 0:
                    ax.text(bar.get_x() + bar.get_width() / 2, height + 0.0003,
                            f'{height:.1%}', ha='center', va='bottom', fontsize=12)
                else:
                    ax.text(bar.get_x() + bar.get_width() / 2, height - 0.0003,
                            f'{height:.1%}', ha='center', va='top', fontsize=12)

    ax.legend(fontsize=14)
    plt.tight_layout()
    return fig

graf_government_consumption_expenditures_and_inv_new = qoq_grouped(government_consumption_expenditures_and_inv, "Government Consumption Expenditures and Gross Investment")
plt.show(graf_government_consumption_expenditures_and_inv_new)
```
